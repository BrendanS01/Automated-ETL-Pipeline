```python
import pandas as pd
import os
import shutil

# Folders (NOT files)
INPUT_FOLDER = "/content/input_folder"
PROCESSED_FOLDER = "/content/processed_folder"
ARCHIVE_FOLDER = "/content/archive_folder"

# Create folders
os.makedirs(INPUT_FOLDER, exist_ok=True)
os.makedirs(PROCESSED_FOLDER, exist_ok=True)
os.makedirs(ARCHIVE_FOLDER, exist_ok=True)

# File path
file_path = os.path.join(INPUT_FOLDER, "data.csv")

# Check if file exists
if not os.path.isfile(file_path):
    print("❌ data.csv not found inside input_folder")
else:
    # 1️⃣ Read file
    df = pd.read_csv(file_path)
    print("File Read Successfully")

    # 2️⃣ Remove duplicates
    df = df.drop_duplicates()

    # 3️⃣ Remove null values
    df = df.dropna()

    # 4️⃣ Standardize date columns
    for col in df.columns:
        if "date" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce")
            df[col] = df[col].dt.strftime("%Y-%m-%d")

    df = df.dropna()

    # 5️⃣ Save cleaned file
    processed_path = os.path.join(PROCESSED_FOLDER, "cleaned_data.csv")
    df.to_csv(processed_path, index=False)

    # 6️⃣ Archive original
    archive_path = os.path.join(ARCHIVE_FOLDER, "data.csv")
    shutil.move(file_path, archive_path)

    print("🎉 ETL Process Completed Successfully!")
```
